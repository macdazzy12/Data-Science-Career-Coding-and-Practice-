```python
import re

# Extract superhero names and groups from the issue descriptions
def extract_superheroes(description):
    # Ensure the description is a string
    if isinstance(description, str):
        # RegEx pattern to find superhero names/groups (e.g., 'X-Men', 'Wolverine', 'Storm')
        superhero_pattern = r'\b(?:X-Men|Wolverine|Storm|Iron Man|Spider-Man|Captain America|Thor|Hulk|Black Panther|Deadpool)\b'
        # Find all matching superheroes/groups in the description
        return re.findall(superhero_pattern, description)
    else:
        # Return an empty list if the description is not a string (e.g., NaN or a number)
        return []

# Apply the extraction function to the 'issue_description' column in your dataset
MarvelComic_data['superheroes'] = MarvelComic_data['issue_description'].apply(extract_superheroes)

# Display the superheroes found in the first few issues
MarvelComic_data[['comic_name', 'superheroes']].head()
```

```python
# RegEx patterns for different topics
topics_patterns = {
    "villains": r'\b(villain|villains|evil|dark)\b',
    "heroic": r'\b(heroic|hero|bravery|courage)\b',
    "adventure": r'\b(adventure|quest|journey|exploration)\b',
    "X-Men": r'\bX-Men\b',
    "Avengers": r'\bAvengers\b'
}
# Function to find topics in the descriptions
def find_topics(description):
    # Convert non-string values (e.g., NaN, float) to empty string
    if not isinstance(description, str):
        description = ""  # If it's not a string, treat it as an empty string
    
    topics_found = {}
    for topic, pattern in topics_patterns.items():
        # Find all occurrences of the topic in the description
        matches = re.findall(pattern, description, re.IGNORECASE)
        topics_found[topic] = len(matches)
    return topics_found
# Apply the function to the 'issue_description' column
MarvelComic_data['topics'] = MarvelComic_data['issue_description'].apply(find_topics)
# Display the topics found in the first few issues
MarvelComic_data[['comic_name', 'topics']].head()
```

```python
# Create a summary of topic frequencies
topic_counts = pd.DataFrame(MarvelComic_data['topics'].tolist())

# Sum up the occurrences of each topic across all descriptions
topic_frequency = topic_counts.sum()

# Display the frequency of each topic
print("Most Frequently Occurring Topics:")
print(topic_frequency.sort_values(ascending=False))
```

```python
# RegEx pattern to extract individual X-Men members
xmen_members_pattern = r'\b(Wolverine|Storm|Cyclops|Jean Grey|Beast|Nightcrawler|Professor X|Rogue|Gambit)\b'

# Function to extract X-Men members from descriptions
def extract_xmen_members(description):
    if isinstance(description, str):
        return re.findall(xmen_members_pattern, description)
    else:
        return []

# Apply the extraction function to the 'issue_description' column
MarvelComic_data['xmen_members'] = MarvelComic_data['issue_description'].apply(extract_xmen_members)

# Display the X-Men members found in the first few issues
MarvelComic_data[['comic_name', 'xmen_members']].head()
```
