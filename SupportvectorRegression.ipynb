```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.inspection import permutation_importance



```

```python
df = pd.read_excel(r"C:\Users\HP\Downloads\Documents\Hydrological Machine Learning Dataset.xlsx")
```

```python
df.head()
```

```python
df.tail()
```

```python
df_clean = df.dropna(subset=['Q 25yr'])
```

```python
features = ['Area(ha)', 'Lov(m)', 'Tc(mins)', 'Runoff Coefficient(C)', 'Sov', 'I 25yr(mm/hr)']
X = df_clean[features]
y = df_clean['Q 25yr']
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svr_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svr_model.predict(X_test_scaled)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)

print("Model Performance Metrics")
print("-------------------------")
print(f"R² Score: {r2:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x=y_test, y=y_pred, color="blue", s=80)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Actual Q 25yr")
plt.ylabel("Predicted Q 25yr")
plt.title("SVR: Actual vs Predicted Q 25yr")
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(df_clean[features + ['Q 25yr']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
results = permutation_importance(
    svr_model, X_test_scaled, y_test, n_repeats=20, random_state=42
)

importance_df = pd.DataFrame({
    "Feature": features,
    "Importance": results.importances_mean
}).sort_values(by="Importance", ascending=False)

print("\nPermutation Importance:\n", importance_df)
plt.figure(figsize=(8,5))
sns.barplot(
    x="Feature", 
    y="Importance", 
    hue="Feature",     
    data=importance_df, 
    palette="viridis", 
    dodge=False, 
    legend=False
)
plt.xticks(rotation=45)  
plt.title("Feature Importance (Permutation - SVR)")
plt.show()


```
